```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the WiderFace dataset txt file which contains bounding box info
with open('./wider_face_split/wider_face_train_bbx_gt.txt', 'r') as f:
    lines = f.readlines()

# print(len(lines))
i = 0
while i < len(lines):
    image_file = lines[i].strip()
    num_faces = int(lines[i+1])
    image_name = image_file.split('/')[-1]
    image_name = image_name.split('.')[0]
    boxes = [list(map(int, line.strip().split())) for line in lines[i+2:i+2+num_faces]]
    
    # Load the image
    img = cv2.imread('WIDER_train/images/' + image_file)
    
    # Crop faces from the image using bounding boxes
    for index, box in enumerate(boxes):
        x, y, w, h = box[:4]
        face = img[y:y+h, x:x+w]
        
        try:
          if face.shape[0] >= 56 and face.shape[1] >= 56:
            face_resized = cv2.resize(face, (112, 112))
            cv2.imwrite(f'./faces_large/{image_name}_{index}.jpg', face_resized)
        except:
          continue
        # print(f'./faces/{image_name}_{num_faces}')
        # Save the cropped face. In this case, faces are saved in a directory named "faces"
        

    i += 2 + num_faces
    if num_faces == 0:
      i += 1
```
